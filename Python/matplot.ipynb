```python
import matplotlib.pyplot as py

x=[1,2,3,4,5]
y=[1,4,9,16,25]
py.xlabel('X-axis')
py.ylabel('Y-axis')
py.title('Squares')
py.plot(x,y,color='red',linestyle='--',marker='o',linewidth=5,markersize=9)
py.grid(True)
```

```python
#MULTIPE PLOT
x=[1,2,3,4,5]
y1=[10,20,30,40,50]
y2=[15,5,35,5,55]

py.figure(figsize=(9,9))

py.subplot(1,2,1)
py.plot(x,y1,color='green')
py.title('Graph 1')

py.subplot(1,2,2)
py.plot(x,y2,color='red')
py.title('Graph 2')

py.subplot(2,2,3)
py.plot(x,y2,color='blue')
py.title('Graph 3')

py.subplot(2,2,4)
py.plot(x,y2,color='yellow')
py.title('Graph 4')
```

```python
#BAR PLOT
cats=['a','b','c','d','e','f']
plots=[1,2,3,4,5,6]
py.bar(cats,plots,color='pink')
```

HISTOGRAMS

```python
histo=[1,1,1,2,3,4,5,5,6,7,5,5,1,31,14,32,4,35,5,41,1]
py.hist(histo,bins=3,color='green',edgecolor='black')
```

```python
#scatter plot
x=[1,2,3,4,5]
y=[1,2,3,4,5]
py.scatter(x,y)
```

```python
#pie chart

label=['a','b','c','d']
size=[10,20,30,40]
colours=['gold','red','green','blue']
xeplode=(0.5,0,0.0,0.1)
py.pie(size,explode=xeplode,labels=label,colors=colours)
```

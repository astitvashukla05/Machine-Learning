```python
import sqlite3

connection=sqlite3.connect('test.db')
connection

```

```python
cursor=connection.cursor()
cursor.execute('''
               create table if not exists emp(
                   id int primary key,
                   name varchar(20) not null,
                   dept varchar(20),
                   age int
                   
               )
               ''')
# commit chagnes
connection.commit()
```

```python
cursor.execute('''
            insert into emp values
            (
                103,"asa","cse",123
            )
        
           ''')

connection.commit()
```

```python
cursor.execute('select * from emp')
row=cursor.fetchall()
for  r in row:
    print(row)
```
